```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
crypto_file_path ="data/crypto_data.csv"
crypto_df = pd.read_csv(crypto_file_path)
crypto_df
```

```python
crypto_df.dtypes
```

```python
crypto_df.convert_dtypes().dtypes
```

```python
crypto_df = crypto_df.loc[crypto_df["IsTrading"] == True]

crypto_df
```

```python
crypto_df = crypto_df.drop(columns=['IsTrading'])
crypto_df
```

```python
crypto_df = crypto_df.dropna()
crypto_df
```

```python
crypto_df = crypto_df.loc[crypto_df["TotalCoinsMined"] > 0]
crypto_df
```

```python
crypto_df = crypto_df.drop(columns=['CoinName', 'Unnamed: 0'])
crypto_df
```

```python
crypto_df = crypto_df.reset_index(drop=True)
crypto_df
```

```python
crypto_learner_df = pd.get_dummies(crypto_df)
crypto_learner_df
```

```python
scaler = StandardScaler().fit(crypto_learner_df)
crypto_scaled = scaler.transform(crypto_learner_df)
crypto_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.90)

crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca.shape
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=200)
tsne_features = tsne.fit_transform(crypto_pca)
```

```python
tsne_features.shape
```

```python
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(tsne_features)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_data)
elbow_df
```

```python
plt.plot(elbow_df['k'], elbow_df['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=5)
model.fit(tsne_features)
predictions = model.predict(tsne_features)
predictions
```

```python
crypto_df["class"] = model.labels_
crypto_df
```
